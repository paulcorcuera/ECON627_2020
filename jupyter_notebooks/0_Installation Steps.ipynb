# Using Julia with Jupyter Notebooks


## Installation Steps 


### Step 1: Install Jupyter 

I strongly advice to install Anaconda to be able to use Jupyter Notebooks (standard installation comes with Python). You can download it from this [link](https://www.anaconda.com/products/individual) and choosing the one that works in your operating system. 

### Step 2: Install Julia 

You can download Julia from the [official site](https://julialang.org/downloads/).

### Step 3: Build Conda Environment for Julia

Open the Julia app and a command prompt will open. 

![img](https://media.geeksforgeeks.org/wp-content/uploads/20200707105720/Screenshot22.png) 


Now we need to install the `IJulia` package. To do so, you can type the following lines and press Enter: 

```julia
    using Pkg
    Pkg.add("IJulia")
```

Wait until Julia installs the package. After it's done the connection between Anaconda and Julia has been established.  



## How to Start Working on your Stuff

You will notice that you have an app named Anaconda Prompt. Open such application and a command prompt will open. If it's possible run this application as administrator.  

### Change directory to the desired workspace 

You only need to type 
```
    cd desired_folder
```

### Open Jupyter Notebooks

Type 
```
    jupyter lab
```

Be patient, and a webpage that looks like this will pop-up 

![img2](https://static.packt-cdn.com/products/9781788998369/graphics/609de416-fc00-4144-b23a-cdf26d45a85e.png)


If you click on the Julia icon you will open a new jupyter notebook that renders Julia code. Otherwise, you can click on `File` and create a new notebook from there.  

### Saving Jupyter Notebooks 

You can of course save your Jupyter notebooks that have the `.ipynb` extension if you click `Save as...`. You can also move jupyter notebooks to your folder and open them (for example, this sample file). I strongly advice to click on `File` and then `Export Notebook as HTML` and send your problem sets under this format. To export as PDF you require to do some extra steps that I will omit in this explanation.
